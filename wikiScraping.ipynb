```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

I choose to treat this scrapping in a jupyter notebook since loading the data once should be enough.

```python
URL = 'https://en.wikipedia.org/wiki/List_of_airline_codes'

page = requests.get(URL)
 
# scrape webpage
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
soup.prettify()
```

```python
tables = soup.findAll("table", { "class" : "wikitable" })
table  = str(tables[0])

table = table.replace("\n",'')
table = table.replace("</tr>", '')
table = table.replace("</td>", '')
table = table.replace("n/a", '')


data  = [r.split('<td>') for r in table.split("<tr>")][2:]

```

```python
df = pd.DataFrame(data)[[2,5]]
df.rename(columns={2:"ICAO",5:"Country"}, inplace=True)

df
```
