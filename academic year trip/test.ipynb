```python
import pandas as pd
import json 
```

```python
with open("data.json") as f :
    data = json.load(f)
    df_eval = pd.DataFrame(data["evaluations"])
    teachers = data["teachers"]
    students = data["student"]

df_eval
```

```python
list_teachers = {"name":[],"surname":[],"classes":[],"subject":[]}
for key,values in teachers.items():
    list_teachers["name"].append(values["name"])
    list_teachers["surname"].append(values["surname"])
    list_teachers["classes"].append(values["classes"])
    list_teachers["subject"].append(values["subject"])
df_teachers = pd.DataFrame(list_teachers)
df_teachers
```

```python
list_students = {"name":[],"surname":[],"class":[]}
for key,values in students.items():
    list_students["name"].append(values["name"])
    list_students["surname"].append(values["surname"])
    list_students["class"].append(values["class"])
df_students = pd.DataFrame(list_students)
df_students
```

# Q1 : For each student, find every subjects on which he has been evaluated.

```python
for i in range(len(df_students.index)):
    print(df_students["surname"].iloc[i], df_students["name"].iloc[i])
    print([x for x in df_eval[df_eval["student"] == str(i)]["subject"].drop_duplicates()])
```

# Q2 For each student, calculate his average in each subject.

```python
for i in range(len(df_students.index)):
    print(df_students["surname"].iloc[i], df_students["name"].iloc[i])
    for subject in df_eval[df_eval["student"] == str(i)]["subject"].drop_duplicates():
        print(subject , " : ",  df_eval[["student","subject","score"]][(df_eval["student"] == str(i)) & (df_eval["subject"] == subject)]["score"].mean())
```

# Q3 Calculate the average of each class in each subject.


```python
list_classes = []
for i in range(len(df_eval.index)):
    list_classes.append(df_students["class"].iloc[int(df_eval.loc[i]["student"])])
df_eval["class"] = list_classes
```

```python
for classe in df_students["class"].drop_duplicates():
    print("Class n°", classe , " : ")
    for subject in df_eval[df_eval["class"] == classe]["subject"].drop_duplicates():
        print(subject, " : ", df_eval[(df_eval["class"] == classe) & (df_eval["subject"] == subject)]["score"].mean())
```

# Q4 For each written evaluations, find all the students that missed the evaluation.


```python
df_eval[df_eval["type"] == "written"].drop_duplicates(subset=["subject","date"])
```

```python
len(df_students[df_students["class"] == 0])
```

```python
df_eval_no_dup = df_eval[df_eval["type"] == "written"].drop_duplicates(subset=["subject","date"])
df_eval_no_dup
```

```python
dic_missed = {"eval":[],"student":[]}

for i in df_eval_no_dup.index:
    if len(df_eval[(df_eval["subject"] == df_eval["subject"].iloc[i]) & (df_eval["date"] == df_eval["date"].iloc[i])])\
        != len(df_students[df_students["class"] == df_eval["class"].iloc[i]]) :
        dic_missed["eval"].append((df_eval["subject"].iloc[i],df_eval["date"].iloc[i]))
        missing_students = []
        for stud in df_students[df_students["class"] == df_eval["class"].iloc[i]].index.tolist() :
            if df_eval[(df_eval["subject"] == df_eval["subject"].iloc[i]) & (df_eval["date"] == df_eval["date"].iloc[i]) & (df_eval["student"] == str(stud))].empty:
                missing_students.append(stud)
        dic_missed["student"].append(missing_students)
```

```python
df_missed = pd.DataFrame(dic_missed)
df_missed
```

# Q5 For each class and for each oral evaluations, find the upper and lower bounds for that evaluation period for the whole class.


```python
df_eval_oral = df_eval[df_eval["type"] == "oral"]
df_eval_oral
```

```python
df_eval.drop_duplicates("subject")
```

```python
eval_bounds = {"eval":[],"class":[],"upper":[],"lower":[]}

for cl in range(5):
    for subj in df_eval["subject"].drop_duplicates() :
        if not df_eval_oral[(df_eval_oral["class"] == cl) & (df_eval_oral["subject"] == subj)].empty:
                eval_bounds["eval"].append(subj)
                eval_bounds["class"].append(cl)
                eval_bounds["lower"].append(df_eval_oral[(df_eval_oral["class"] == cl) & (df_eval_oral["subject"] == subj)].min()["score"])
                eval_bounds["upper"].append(df_eval_oral[(df_eval_oral["class"] == cl) & (df_eval_oral["subject"] == subj)].max()["score"])

df_bounds = pd.DataFrame(eval_bounds)
df_bounds
```

# Q6 Create an object representing a class with relevant field/members.

```python
class Classe():

    def __init__(self, id, list_students, list_teachers, list_written_eval, list_oral_eval) -> None:
            id = id
            list_students = list_students
            list_teachers = list_teachers
            list_written_eval = list_written_eval
            list_oral_eval = list_oral_eval

    def __str__(self):
        return f"The class n°{self.id} has {len(self.list_students)} students and {len(self.list_teachers)}"
```
