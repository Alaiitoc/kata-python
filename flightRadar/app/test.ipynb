```python
df_airports = pd.DataFrame(fr_api.get_airports())
df_airports
```

```python
def in_continent(x):
    lat,lon = x['lat'],x['lon']
    zones = fr_api.get_zones()
    for continent,item in zones.items():
        if item["br_y"] < lat < item["tl_y"] and item["tl_x"] < lon < item["br_x"] :
            return 'africa' if continent == 'maldives' else continent
    return None
```

```python
in_continent(df_airports.iloc[0]['lat'],df_airports.iloc[0]['lon'])
```

```python
df_airports["Continent"] = df_airports[['lat','lon']].apply(in_continent,axis=1)
```
